```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import matplotlib
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

import FinanceDataReader as fdr
```

```python
df = pd.read_csv("hsgdata/big3_1.csv")
df
```

```python
matplotlib.rcParams['font.family'] ='Malgun Gothic' # 폰트설정
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스기호(-)가 깨지는 현상 방지

plt.figure(figsize=(10,5))
plt.plot(df["year"],df["kyobo"], color="red", linewidth = 2,label = "교보문고",marker='o')
plt.title("'BIG3' 인터넷서점 매출 추이", color = "#5C6BC0", fontsize = 25, fontweight = "bold")
plt.xticks(rotation = 30)
plt.ylabel("단위 : 억원", loc = "top", color="black")
plt.grid(axis='y')
plt.plot(df["year"],df["yes24"],label = "YES24", color="green",linewidth = 2,marker='o')
plt.plot(df["year"],df["aladin"],label = "알라딘", marker='o', color="orange",linewidth = 2)
plt.legend(loc = "upper left")
plt.savefig("hsgdata/빅3.png")
```
