```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
df = pd.read_csv("hsgdata/yes24store.csv")
df
```

```python
matplotlib.rcParams['font.family'] ='Malgun Gothic' # 폰트설정
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스기호(-)가 깨지는 현상 방지

plt.figure(figsize=(10,5))
plt.plot(df["year"],df["yes24"], color="orange", label="yes24 매출",linewidth = 2,marker='o')
plt.title("도서정가제 시행 전과 시행 후", color = "#5C6BC0", fontsize = 25, fontweight = "bold")
plt.xticks(rotation = 30)
plt.grid(axis='y')
plt.plot(df["year"],df["bookstore"], label="독립서점 수",color="green",linewidth = 2,marker='o')
plt.axvline(x=2013, color='r', linewidth=2, label="도서정가제 시행")
plt.legend(loc = "upper left")
plt.savefig("hsgdata/yes24vsbookstore.png")
```
